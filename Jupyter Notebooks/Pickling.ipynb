```python
import datetime
#df_original.drop(['elevgain', 'startdate', 'weight', 'sex', 'year', 'cumulative_elev_loss', 'cumulative_elev_gain', 'heartrate', 'partitionid'], axis = 1)
#Take the needed columns
new = df_original[['hashedathleteid', 'startdatelocal', 'totaldistance', 'time_final', 'pace_diff', 'pace_cumul']].copy()
#Get rid of false distances
new = new[new.totaldistance > 50]
new = new[new.totaldistance < 50000]
#Get rid of activities after the marathon date
x = datetime.datetime(2016, 11, 1)
new = new[new.startdatelocal < x]
new.sort_values(by=['startdatelocal'], inplace=True)
new = new.reset_index(drop=True)
new
```
